# Graphing/Plotting for results 

### Q1. Is Dynamic/Static model better for CI/NH 

```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 
```

```python
model_data = pd.read_csv("Findings/dimension_model_results.csv")

model_data
```

```python
#group by language model and group 
model_groups = model_data.groupby(["language_model", "group", "model"])

#first graph 
w2v_NH_dynamic = model_groups.get_group(("word2vec", "NH", "Dynamic/Simdrop"))
w2v_NH_static = model_groups.get_group(("word2vec", "NH", "Static/None"))

#second graph 
w2v_CI_dynamic = model_groups.get_group(("word2vec", "CI", "Dynamic/Simdrop"))
w2v_CI_static = model_groups.get_group(("word2vec", 'CI', 'Static/None'))

#third graph 
s2v_NH_dynamic = model_groups.get_group(("speech2vec", "NH", "Dynamic/Simdrop"))
s2v_NH_static = model_groups.get_group(("speech2vec", "NH", "Static/None"))

#fourth graph 
s2v_CI_dynamic = model_groups.get_group(("speech2vec", "CI", "Dynamic/Simdrop"))
s2v_CI_static = model_groups.get_group(("speech2vec", 'CI', 'Static/None'))




```

```python
'''Word2Vec NH graph'''
plt.scatter(w2v_NH_dynamic['dimension'], w2v_NH_dynamic['sum_NLL'], c='r', label = "Dynamic")
plt.plot(w2v_NH_dynamic['dimension'], w2v_NH_dynamic['sum_NLL'], c='r')
plt.scatter(w2v_NH_static['dimension'], w2v_NH_static['sum_NLL'], c="b", label = "Static")
plt.plot(w2v_NH_static['dimension'], w2v_NH_static['sum_NLL'], c="b")
plt.title("Word2Vec NH Dynamic/Static")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q1/word2vec_NH")
plt.show()

```

```python
'''Word2Vec CI Graph'''
plt.scatter(w2v_CI_dynamic['dimension'], w2v_CI_dynamic['sum_NLL'], c='r', label = "Dynamic")
plt.plot(w2v_CI_dynamic['dimension'], w2v_CI_dynamic['sum_NLL'], c='r')
plt.scatter(w2v_CI_static['dimension'], w2v_CI_static['sum_NLL'], c="b", label = "Static")
plt.plot(w2v_CI_static['dimension'], w2v_CI_static['sum_NLL'], c="b")
plt.title("Word2Vec CI Dynamic/Static")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q1/word2vec_CI")
plt.show()
```

```python
'''Speech2Vec NH Graph'''
plt.scatter(s2v_NH_dynamic['dimension'], s2v_NH_dynamic['sum_NLL'], c='r', label = "Dynamic")
plt.plot(s2v_NH_dynamic['dimension'], s2v_NH_dynamic['sum_NLL'], c='r')
plt.scatter(s2v_NH_static['dimension'], s2v_NH_static['sum_NLL'], c="b", label = "Static")
plt.plot(s2v_NH_static['dimension'], s2v_NH_static['sum_NLL'], c="b")
plt.title("Speech2Vec NH Dynamic/Static")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q1/speech2vec_NH")
plt.show()
```

```python
'''Speech2Vec CI Graph'''
plt.scatter(s2v_CI_dynamic['dimension'], s2v_CI_dynamic['sum_NLL'], c='r', label = "Dynamic")
plt.plot(s2v_CI_dynamic['dimension'], s2v_CI_dynamic['sum_NLL'], c='r')
plt.scatter(s2v_CI_static['dimension'], s2v_CI_static['sum_NLL'], c="b", label = "Static")
plt.plot(s2v_NH_static['dimension'], s2v_CI_static['sum_NLL'], c="b")
plt.title("Speech2Vec CI Dynamic/Static")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q1/speech2vec_CI")
plt.show()

```

### Q2. Speech2vec (speech + text data), word2vec (text data), which language model fits better with NH and CI

```python
model_data = pd.read_csv("Findings/dimension_model_results.csv")

model_data
```

```python
#group by language model and group 
model_groups = model_data.groupby(["language_model", "group", "model"])

#first graph 
w2v_NH_dynamic = model_groups.get_group(("word2vec", "NH", "Dynamic/Simdrop"))
s2v_NH_dynamic = model_groups.get_group(("speech2vec", "NH", "Dynamic/Simdrop"))

#second graph 
w2v_CI_dynamic = model_groups.get_group(("word2vec", "CI", "Dynamic/Simdrop"))
s2v_CI_dynamic = model_groups.get_group(("speech2vec", "CI", "Dynamic/Simdrop"))
```

```python
'''NH dynamic'''
plt.scatter(w2v_NH_dynamic['dimension'], w2v_NH_dynamic['sum_NLL'], c='r', label = "Word2Vec")
plt.plot(w2v_NH_dynamic['dimension'], w2v_NH_dynamic['sum_NLL'], c='r')
plt.scatter(s2v_NH_dynamic['dimension'], s2v_NH_dynamic['sum_NLL'], c='b', label = "Speech2Vec")
plt.plot(s2v_NH_dynamic['dimension'], s2v_NH_dynamic['sum_NLL'], c='b')
plt.title("NH Language Models")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q2/NH_language_models")
plt.show()

```

```python
'''CI dynamic'''
plt.scatter(w2v_CI_dynamic['dimension'], w2v_CI_dynamic['sum_NLL'], c='r', label = "Word2Vec")
plt.plot(w2v_CI_dynamic['dimension'], w2v_CI_dynamic['sum_NLL'], c='r')
plt.scatter(s2v_CI_dynamic['dimension'], s2v_CI_dynamic['sum_NLL'], c='b', label = "Speech2Vec")
plt.plot(s2v_CI_dynamic['dimension'], s2v_CI_dynamic['sum_NLL'], c='b')
plt.title("CI Language Models")
plt.xlabel("Dimensions")
plt.ylabel("NLL")
plt.legend(loc="lower right")
plt.savefig("Findings/Figures/Q2/CI_language_models")
plt.show()

```
