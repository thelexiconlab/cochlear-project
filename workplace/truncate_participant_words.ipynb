```python
import pandas as pd
import numpy as np
```

```python
participant_truncation = pd.read_csv("analysis/outputs/length_matched_items.csv")

old_participant_truncation = pd.read_csv("analysis/outputs/length_matched_items.csv")
old_part_dict = [[x, y] for x, y in zip(old_participant_truncation['ID_NH'], old_participant_truncation['ID_CI'])]


rep_data = pd.read_csv('forager/data/fluency_lists/participant_data/transformed-data.csv')
no_rep_data = pd.read_csv("forager/data/fluency_lists/participant_data/no_rep_data.csv")
participant_order = pd.Series(no_rep_data['ID']).drop_duplicates().tolist()
new_participant_truncation = pd.DataFrame(columns=['NH', 'ID_NH', 'n_NH', 'CI', 'ID_CI', 'n_CI', 'n_final'])

for part in old_part_dict: 
    df = no_rep_data.groupby('ID')
    NH = part[0]
    CI = part[1]
    row = ['NH', NH, len(df.get_group(NH)), 'CI', CI, len(df.get_group(CI)), min(len(df.get_group(NH)), len(df.get_group(CI)))]
    new_participant_truncation.loc[len(new_participant_truncation)] = row
    
new_participant_truncation.to_csv('analysis/outputs/new_length_matched_items.csv', index=False)


```

```python
participant_truncation = pd.read_csv("analysis/outputs/new_length_matched_items.csv")
participant_truncation
```

```python
NH_participants = participant_truncation['ID_NH'].tolist()
CI_participants = participant_truncation['ID_CI'].tolist()
n_final = participant_truncation['n_final'].tolist()
```

```python
trunc_dict = {} 
for i in range(len(NH_participants)):
    trunc_dict[NH_participants[i]] = n_final[i]
    trunc_dict[CI_participants[i]] = n_final[i]
print(trunc_dict)
```

```python
trunc_df = pd.DataFrame(columns=['ID', 'Word'])
trunc_df
```

```python
participant_order = pd.Series(no_rep_data['ID']).drop_duplicates().tolist()
participant_order
```

```python
for participant in participant_order: 
    df = no_rep_data.groupby(['ID']).get_group(participant).reset_index(drop=True)

    for i in range(trunc_dict[participant]):
        trunc_df.loc[len(trunc_df)] = df.loc[i]
trunc_df.to_csv('forager/data/fluency_lists/participant_data/truncated_data.csv', index=False)
```
