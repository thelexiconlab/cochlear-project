```python
import pandas as pd 
# gotta match 1114 
no_rep = pd.read_csv("no_rep_data.txt", sep="\t")
no_rep

participant_full_dict = pd.read_csv("new_length_matched_items.csv")
original_word_count = {}
pair = []
for i in range(len(participant_full_dict)):
    # getting pairs for count matching
    pair.append([participant_full_dict['ID_NH'][i], participant_full_dict['ID_CI'][i]])
    
    # get original word count
    grouping = no_rep.groupby('ID')
    original_word_count[participant_full_dict['ID_NH'][i]] = len(grouping.get_group(participant_full_dict['ID_NH'][i]))
    original_word_count[participant_full_dict['ID_CI'][i]] = len(grouping.get_group(participant_full_dict['ID_CI'][i]))

    
```

```python
word_dict = {} 
num_dict = {} 

for i in range(len(participant_full_dict)): 
    grouping = no_rep.groupby('ID')
    participant1 = grouping.get_group(pair[i][0])
    participant1
    participant2 = grouping.get_group(pair[i][1])
    
    if len(participant1) > len(participant2):
        word_dict[pair[i][0]] = participant1['Word'].tolist()[:len(participant2)]
        word_dict[pair[i][1]] = participant2['Word'].tolist()
        
        num_dict[pair[i][0]] = len(participant2)
        num_dict[pair[i][1]] = len(participant2)
    else: 
        word_dict[pair[i][0]] = participant1['Word'].tolist()
        word_dict[pair[i][1]] = participant2['Word'].tolist()[:len(participant1)]
        
        num_dict[pair[i][0]] = len(participant1)
        num_dict[pair[i][1]] = len(participant1)

```

```python
ids = sorted(set(no_rep['ID'].tolist()))
ids
```

```python
final_df = pd.DataFrame(columns=['ID', 'Word'])
for i in ids: 
    final_df = pd.concat([final_df, pd.DataFrame({'ID': i, 'Word': word_dict[i]})])
final_df
```
